```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import sys
sys.path.append("../library-Git/")
from defaultPlot import *
from stringFormat import *

```

```python
def readDist(t_fileName):
    data = pd.read_csv(t_fileName, sep=",", header=None)
    data = data.values.transpose()
    return tuple([row for row in data])

def readSIS(t_fileName):
    data = pd.read_csv(t_fileName, sep="\t", header=None)
    data = data.values.transpose()
    return tuple([row for row in data])
```

```python
_, ax = plt.subplots(figsize = (10,10))
degree, dist = readDist("../playground/degreeDist_bin.txt")
fitX, fitY, alpha, _ = logFit(degree, dist)

ax.plot(degree, dist, 'o-')
ax.plot(fitX, fitY, 'k-', label = latexString("\\alpha=") + latexFloat(alpha,2))

ax.set_xscale('log'); ax.set_yscale('log')
ax.set_xlabel(latexString("k")); ax.set_ylabel(latexString("P_k"))
ax.legend(loc="upper right")
plt.show()
```

```python
_, ax = plt.subplots(figsize = (10,10))
degree, dist = readDist("../playground/nodeWeightDist_bin.txt")
fitX, fitY, alpha, _ = logFit(degree, dist)

ax.plot(degree, dist, 'o-')
ax.plot(fitX, fitY, 'k-', label = latexString("\\alpha=") + latexFloat(alpha,2))

ax.set_xscale('log'); ax.set_yscale('log')
ax.set_xlabel(latexString("n")); ax.set_ylabel(latexString("P_n"))
ax.legend(loc="upper right")
plt.show()
```

```python
_, ax = plt.subplots(figsize = (10,10))
ratioS, ratioI = readSIS("../data/epidemics/SIS_meta/N1.0e+02,M4,SIII8.00,IS2.00-0.txt")
t = np.linspace(0.0, len(ratioS) * 0.1 - 0.1, len(ratioS))

ax.plot(t,ratioS, 'o-', label="S")
ax.plot(t,ratioI,'o-', label="I")

ax.set_xlabel(latexString("t")); ax.set_ylabel(latexString("\\rho"))
ax.legend(loc="upper right")
plt.show()
```
